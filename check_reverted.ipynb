```python
from github import Github

# =====================
# CONFIGURATION
# =====================
TOKEN = "" # Replace with your GitHub Personal Access Token
REPO_NAME = "facebook/react" 
PR_NUMBER = 32603
# MAIN_BRANCH is now determined automatically
# =====================

# Authenticate to GitHub
try:
    g = Github(TOKEN)
    repo = g.get_repo(REPO_NAME)
    pr = repo.get_pull(PR_NUMBER)
    
    # ðŸŒŸ AUTOMATIC BRANCH DETECTION ðŸŒŸ
    # Get the repository's default branch (which is usually the main/primary branch)
    MAIN_BRANCH = repo.default_branch 
    print(f"âœ… Automatically determined main branch: **{MAIN_BRANCH}**")
    
except Exception as e:
    print(f"âŒ Error during GitHub authentication or repository access: {e}")
    exit()

# --- Core Logic to Check for Revert ---

def check_pr_revert(pr, repo, main_branch_name):
    """Checks if a merged PR has been reverted."""
    
    # 1. Ensure the PR was actually merged
    if not pr.merged:
        return f"âŒ PR #{pr.number} has not been merged yet."

    merge_commit_sha = pr.merge_commit_sha
    if not merge_commit_sha:
        return f"âŒ PR #{pr.number} is merged but is missing a merge commit SHA (usually due to a non-merge merge method)."
    
    # 2. Get the main branch history
    try:
        # Note: This check is mostly for verification; the name is now from repo.default_branch
        repo.get_branch(main_branch_name) 
    except Exception:
        # This error is now highly unlikely if repo.default_branch is used correctly
        return f"âŒ Error: Default branch '{main_branch_name}' could not be accessed."

    # 3. Search for a revert commit in the main branch history
    # Git revert commits usually contain the pattern: 'This reverts commit <SHA>'
    revert_message_fragment = f"reverts commit {merge_commit_sha[:10]}" # Search for first 10 characters
    
    # We search the commits on the main branch for the specific revert message
    commits = repo.get_commits(sha=main_branch_name)
    
    for commit in commits:
        if commit.commit.message and revert_message_fragment in commit.commit.message:
            # Found the revert commit
            revert_commit_sha = commit.sha
            output = f"âœ… PR #{pr.number} was **REVERTED** by commit **{revert_commit_sha[:7]}**.\n"
            output += f"ðŸ”— Commit URL: {commit.html_url}\n"

            # Optional: Find the Pull Request that merged the revert (if it exists)
            try:
                # Limiting the search to recent closed PRs for performance
                for revert_pr in repo.get_pulls(state='closed', base=main_branch_name, sort='updated', per_page=30):
                    # The revert commit is typically the *last* commit in the revert PR
                    if revert_pr.merge_commit_sha == revert_commit_sha:
                        output += f"ðŸ” Revert was merged via PR **#{revert_pr.number}** - {revert_pr.title}\n"
                        output += f"URL: {revert_pr.html_url}"
                        break
            except Exception:
                # Silently fail if we can't find the associated PR
                pass
            
            return output

    # 4. If the loop finishes without finding a match
    return f"âŒ PR #{pr.number} was merged, but a revert commit was not found in the recent history of the **{main_branch_name}** branch."

# Execute the check
result = check_pr_revert(pr, repo, MAIN_BRANCH)
print("--- Check Result ---")
print(result)
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
Batch, cache-aware PR revert scanner for huge datasets.

Reads:  hf://datasets/hao-li/AIDev/all_pull_request.parquet
Writes: pr_reverts_master.csv (append/resume)
Cache:  .cache_reverts/<owner_repo>/<YYYY-MM>.pkl  (commit messages in windows)

Strategy:
- Group by repo and PR merge month.
- For each group, fetch default-branch commits in [merge_month_start .. merge_month_end + WINDOW_DAYS].
- Mark a PR as reverted if any commit message in that window contains:
    "This reverts commit <merge_sha[:10]>"
- Fallback (no merge_sha): look for commit messages containing "revert" and "#<PR_NUMBER>" (heuristic).

Requirements:
  pip install pandas pyarrow "fsspec[http]" huggingface_hub PyGithub python-dateutil
Auth:
  export GITHUB_TOKEN=ghp_xxx
"""

import os
import re
import time
import json
import math
import pickle
import hashlib
from collections import defaultdict
from datetime import datetime, timedelta, timezone

import pandas as pd
from dateutil import parser as dtparser
from github import Github, GithubException, RateLimitExceededException

# ---------------- Config ----------------
INPUT_PARQUET = "hf://datasets/hao-li/AIDev/all_pull_request.parquet"
OUTPUT_CSV = "pr_reverts_master.csv"
CACHE_DIR = ".cache_reverts"
WINDOW_DAYS_AFTER_MERGE_MONTH = 120   # commits window after the PR merge month
CHUNK_SIZE = 50_000                   # process dataframe in chunks
API_PAUSE_SECS = 0.15                 # polite delay between API ops
ROW_LIMIT = None                      # set e.g. 200_000 to cap for trial runs

# -------------- Regex helpers ----------
REPO_RE = re.compile(r"https?://api\.github\.com/repos/([^/]+/[^/]+)")
PR_URL_RE = re.compile(r"https?://github\.com/([^/]+/[^/]+)/pull/(\d+)")

def parse_repo_full_name(repo_url: str):
    if not isinstance(repo_url, str):
        return None
    m = REPO_RE.match(repo_url.strip())
    return m.group(1) if m else None

def parse_pr_number(html_url: str):
    if not isinstance(html_url, str):
        return None
    m = PR_URL_RE.match(html_url.strip())
    return int(m.group(2)) if m else None

def ensure_dir(p):
    os.makedirs(p, exist_ok=True)

def cache_path(owner_repo: str, month_label: str, window_days: int):
    d = os.path.join(CACHE_DIR, owner_repo.replace("/", "__"))
    ensure_dir(d)
    fname = f"{month_label}__wd{window_days}.pkl"
    return os.path.join(d, fname)

def load_cache(path):
    if os.path.exists(path):
        with open(path, "rb") as f:
            return pickle.load(f)
    return None

def save_cache(path, obj):
    with open(path, "wb") as f:
        pickle.dump(obj, f, protocol=pickle.HIGHEST_PROTOCOL)

def safe_get_repo(g, full_name: str):
    for delay in (0, 2, 4, 8):
        try:
            return g.get_repo(full_name)
        except RateLimitExceededException:
            time.sleep(delay or 2)
        except GithubException as e:
            # 404, 451, or no access
            raise e
    return None

def iso_to_dt(s):
    if pd.isna(s) or not isinstance(s, str):
        return None
    try:
        return dtparser.isoparse(s).astimezone(timezone.utc)
    except Exception:
        return None

def month_bounds(dt_utc: datetime):
    start = dt_utc.replace(day=1, hour=0, minute=0, second=0, microsecond=0)
    # next month start
    if start.month == 12:
        next_start = start.replace(year=start.year+1, month=1)
    else:
        next_start = start.replace(month=start.month+1)
    end = next_start - timedelta(seconds=1)
    return start, end

def fetch_default_branch_commits(repo, since_dt: datetime, until_dt: datetime):
    """
    Fetch commit messages on default branch in [since_dt, until_dt].
    Returns a list of normalized lowercased messages.
    """
    messages = []
    default_branch = repo.default_branch
    # NOTE: PyGithub lets you pass since/until
    commits = repo.get_commits(sha=default_branch, since=since_dt, until=until_dt)
    for c in commits:
        msg = (c.commit.message or "").strip().lower()
        if msg:
            messages.append(msg)
    return messages

def window_for_month(merge_month_start: datetime, merge_month_end: datetime, window_days: int):
    since_dt = merge_month_start
    until_dt = (merge_month_end + timedelta(days=window_days)).replace(tzinfo=timezone.utc)
    return since_dt, until_dt

def build_messages_cache_for_group(github_client, owner_repo, merged_month, repo_obj):
    """
    Build/return cached list of commit messages for (repo, merged_month).
    """
    cache_file = cache_path(owner_repo, merged_month.strftime("%Y-%m"), WINDOW_DAYS_AFTER_MERGE_MONTH)
    cached = load_cache(cache_file)
    if cached is not None:
        return cached

    # compute time window
    mstart, mend = month_bounds(merged_month)
    since_dt, until_dt = window_for_month(mstart, mend, WINDOW_DAYS_AFTER_MERGE_MONTH)

    msgs = fetch_default_branch_commits(repo_obj, since_dt, until_dt)
    save_cache(cache_file, msgs)
    return msgs

def detect_revert_for_row(msgs_lower_list, merge_sha: str, pr_number: int):
    """
    Primary: look for 'reverts commit <sha10>'.
    Fallback: if no merge_sha, or not found, look for 'revert' + '#<pr>'.
    """
    # primary
    if merge_sha:
        needle = f"reverts commit {merge_sha[:10]}".lower()
        for m in msgs_lower_list:
            if needle in m:
                return 1
    # fallback heuristic
    if pr_number:
        fallback_a = f"revert"
        fallback_b = f"#{pr_number}"
        for m in msgs_lower_list:
            if fallback_a in m and fallback_b in m:
                return 1
    return 0

def main():
    token = os.getenv("GITHUB_TOKEN", "").strip()
    if not token:
        raise SystemExit("ERROR: set GITHUB_TOKEN first.")

    g = Github(token)

    print(f"ðŸ“¥ Loading Parquet: {INPUT_PARQUET}")
    df = pd.read_parquet(INPUT_PARQUET, columns=[
        "id","number","repo_url","html_url","merged_at","state"
    ])

    if ROW_LIMIT:
        df = df.head(ROW_LIMIT)

    # Prep minimal frame
    df["repo"] = df["repo_url"].map(parse_repo_full_name)
    df["pr_number"] = df["html_url"].map(parse_pr_number).fillna(df["number"]).astype("Int64")
    df["merged_dt"] = df["merged_at"].map(iso_to_dt)
    df["merged_ym"] = df["merged_dt"].dt.to_period("M").astype("datetime64[ns]")  # month bucket

    # Only consider PRs that are merged (state == closed AND merged_at not null)
    merged_mask = df["merged_dt"].notna()
    work = df.loc[merged_mask, ["id","pr_number","repo","html_url","merged_dt","merged_ym"]].copy()

    # Prepare output file (append/resume)
    out_exists = os.path.exists(OUTPUT_CSV)
    if not out_exists:
        pd.DataFrame(columns=["id","number","repo","html_url","reverted"]).to_csv(OUTPUT_CSV, index=False)

    # Iterate by repo to minimize API context switches
    for repo_name, repo_group in work.groupby("repo", dropna=True):
        if not isinstance(repo_name, str):
            continue
        print(f"\nðŸ·ï¸ Repo: {repo_name}  (rows={len(repo_group)})")

        # Try getting the repo once
        try:
            repo_obj = safe_get_repo(g, repo_name)
        except GithubException as ge:
            print(f"  âš ï¸ Cannot access repo {repo_name}: {ge.status}. Skipping.")
            continue
        if repo_obj is None:
            print(f"  âš ï¸ Repo {repo_name} not accessible. Skipping.")
            continue

        # Process by merged month
        for merged_month, month_group in repo_group.groupby("merged_ym", dropna=True):
            month_dt = pd.Timestamp(merged_month).to_pydatetime().replace(tzinfo=timezone.utc)
            print(f"  ðŸ“… Month: {month_dt.strftime('%Y-%m')}  (n={len(month_group)})")

            # Load or build commit message cache for this (repo, month)
            msgs = build_messages_cache_for_group(g, repo_name, month_dt, repo_obj)

            # For PR-level merge SHA, we need 1 call per PR to fetch merge_commit_sha.
            # We will do it lazily and cache in memory for this month batch.
            reverted_rows = []
            for _, r in month_group.iterrows():
                prn = int(r["pr_number"]) if pd.notna(r["pr_number"]) else None

                # Try to fetch PR to get merge_commit_sha (only once per PR)
                merge_sha = None
                try:
                    pr_obj = repo_obj.get_pull(prn)
                    # merged only set if merged; we filtered that already
                    merge_sha = pr_obj.merge_commit_sha or None
                except GithubException as ge:
                    # 404 or no access; fallback heuristic only
                    pass

                rev = detect_revert_for_row(msgs, merge_sha, prn)
                reverted_rows.append({
                    "id": int(r["id"]),
                    "number": prn,
                    "repo": repo_name,
                    "html_url": r["html_url"],
                    "reverted": rev
                })
                time.sleep(API_PAUSE_SECS)

            # Append this monthâ€™s results immediately (resume-safe)
            pd.DataFrame(reverted_rows)[["id","number","repo","html_url","reverted"]].to_csv(
                OUTPUT_CSV, mode="a", header=not out_exists and os.path.getsize(OUTPUT_CSV) == 0, index=False
            )
            out_exists = True

    print(f"\nâœ… Done. Results appended to {OUTPUT_CSV}")

if __name__ == "__main__":
    main()
```
